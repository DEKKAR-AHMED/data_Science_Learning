# Distribution plots 


```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.distplot(tips["total_bill"],kde=False) # historgram
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips) # between two variable 
```

```python
sns.pairplot(tips,hue="sex",palette="coolwarm") # all varialble relation 
```

```python
sns.rugplot(tips["total_bill"])
```

 # Categorial plots 

```python
import numpy as np

sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.std)
```

```python
# count plot 
sns.countplot(x="sex",data=tips) # cout accurance 
```

```python
#boxplot 
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",split=True)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=True)
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips)
```

```python
sns.factorplot(x="day",y="total_bill",data=tips,kind="violin")
```

# Matrix plot 

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap="coolwarm")
```
