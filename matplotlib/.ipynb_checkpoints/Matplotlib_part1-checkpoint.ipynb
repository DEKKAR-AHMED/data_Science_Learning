```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# to use the plot in jupiter 
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

# function methode 

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,"r-") 
#color
```

```python
plt.xlabel(" X label")
plt.ylabel("Y label")
plt.title("Title")
plt.plot(x,y,"r-") 
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")
plt.subplot(1,2,2)
plt.plot(y,x,"b")
plt.subplot
```

# object oriented methode 

```python
fig = plt.figure()
#figure object 
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("Xlabel")
axes.set_ylabel("lYabel")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.2,0.4,0.2])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title("Bigger Plot")
axes2.set_title("Smaller Plot")
# figue add axxes (postionLargeur,positionHauteur,taillelargeur,taillelongeur)
```

# MATPLOTLIB PART 2 

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
axes[0].set_title("first")
axes[1].set_title("second")
# at the end to avoid the overlaps 
plt.tight_layout()
```

# Figure Size and DPI

```python
'''fig = plt.figure(figsize=(8,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)'''
fig,axes = plt.subplots(2,1,figsize=(8,4))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

# SAVE A FIGURE 

```python
fig.savefig("my_figure.png",dpi=200)
```

```python
# add legend 
fig = plt.figure(figsize=(8,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X squared")
ax.plot(x,x**3,label="X cubed")
ax.legend(loc=0)
```
